## Shapely tutorial

```python
from shapely import MultiPolygon, Polygon
ob = MultiPolygon([
    (
    ((0.0, 0.0), (0.0, 1.0), (1.0, 1.0), (1.0, 0.0)),
    [((0.3,0.3), (0.3,0.4), (0.4,0.4), (0.4,0.3))]
    )
])

print(len(ob.geoms))
print(type(ob.geoms[0]) == Polygon)
print(ob)
```

## Train and validation plots

```python
import pandas as pd


def plot_train_val(fname):
    df = pd.read_csv(fname, sep='\t')
    print(df.head())
    df.plot(x='epoch', y=['train_loss', 'val_loss'], figsize=(10, 10))
    df.plot(x='epoch', y=['train_error_rate', 'val_error_rate'], figsize=(10, 10))


plot_train_val('/Users/cristianion/Desktop/visual_recognition_train/models/inria/inria_model_1_0_1_val.tsv')
```
